```python
import pandas as pd
crashes = pd.read_csv('Traffic_Crashes_-_Crashes.csv', index_col='CRASH_DATE', parse_dates=True)
crashes.head()
```

```python
crashes.head()
```

```python
# daylight crashes
myColumns = crashes[['LATITUDE', 'LONGITUDE', 'LIGHTING_CONDITION', 'INJURIES_FATAL']]
daylightCrashes = myColumns.loc[(myColumns.INJURIES_FATAL > 0) & (myColumns.LIGHTING_CONDITION == 'DAYLIGHT')]
daylightCrashes.to_csv('daylightCrashes.csv', index=True, header=True)
daylightCrashes
```

```python
# darkness crashes
darknessCrashes = myColumns.loc[(myColumns.INJURIES_FATAL > 0) & ((myColumns.LIGHTING_CONDITION == 'DARKNESS, LIGHTED ROAD') | (myColumns.LIGHTING_CONDITION == 'DARKNESS'))]
darknessCrashes.to_csv('darknessCrashes.csv', index=True, header=True)
darknessCrashes
```
